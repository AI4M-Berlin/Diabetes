Let's first filter on features related to the study: "transcranial doppler of cerebral arteries, 24-h blood pressure numerics, high resolution waveforms (ECG, blood pressure, CO2 and respiration) during various movement tasks, 24-h ECG, EMG, and accelerometer recordings, and gait pressure recordings during a walking test."
- (24Hour), StandEC, SitEC, StandEO, SUPINE REBREATHING, " ", etc
- different test: (^HVLT)|(^VF)|(^RCFT)|(^DS)|(Score)'

```python
study_measures = '(24Hour)|(StandEC)|(SitEC)|(StandEO)|(SitEO)|(SUPINE REBREATHING)|((?i)baseline)|((?i)TILT)|(HV)|(Clinical-WMHs)|(^HRV)|(^GAIT)|(^\s)'
list(df.loc[:, df.columns.str.contains(study_measures)].columns)
```

```python
pd.unique(df_desc["Variable description"].where(df_desc["SUBJECT NUMBER"].isin(list(df.loc[:, df.columns.str.contains(study_measures)].columns))))
```

```python
scores_test = '(^HVLT)|(^VF)|(^RCFT)|(^DS)|(Score)'
list(df.loc[:, df.columns.str.contains(scores_test)].columns)
```

```python
f'{study_measures}|{scores_test}'
```

```python
filter_1 = f'{study_measures}|{scores_test}'
df_v1 = df.loc[:, ~df.columns.str.contains(filter_1)]
list(df_v1.columns)
```

```python
df_desc[df_desc["SUBJECT NUMBER"].str.contains("GAIT", na=False)]
```

```python
df_v1.shape
```

```python
missing_values_counts_v1 = df_v1.isna().sum().sort_values(ascending=False)
missing_values_counts_v1.shape #checking
```

```python
fig = px.bar(missing_values_counts_v1, height=600, width=2000, labels={'index': 'features'})
fig.update_layout(font_size=8, bargap=0.3)
fig.show()
```

There are 88 patients, we can filter on features with more than 57 missing values (1/3 of the datasets) The threshold is volontary high
- to keep as much features as possible
- adjust with the future merging of other dataset

```python
len(df_v1)*1/3
```

```python
df_v1.shape
```

```python
(missing_values_counts_v1 > len(df_v1)*1/3).sum()
# 59 features are missing more than 57 values
```

```python
df_v2 = df_v1.dropna(thresh=(len(df_v1)*2/3), axis='columns')
```

```python
missing_values_counts_v2 = df_v2.isna().sum().sort_values(ascending=False)
missing_values_counts_v2.shape #checking
```

```python
fig = px.bar(missing_values_counts_v2, height=600, width=1500, labels={'index95': 'features'})
fig.update_layout(font_size=10, bargroupgap=0.3)
fig.show()
```
