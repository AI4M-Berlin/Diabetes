## Common elements

```python
df_desc_v1 = df_desc.copy()
df_desc_v1["SUBJECT NUMBER"] = df_desc_v1["SUBJECT NUMBER"].str.lower()
print(list(df_desc_v1["SUBJECT NUMBER"])[0:10])
df_stroke_v2 = df_stroke_v1.copy() 
df_stroke_v2.columns = df_stroke_v1.columns.str.lower()
print(df_stroke_v2.columns)
df_perfusion_v2 = df_perfusion_v1.copy() 
df_perfusion_v2.columns = df_perfusion_v1.columns.str.lower()
print(df_perfusion_v2.columns)
```

```python
common_col = set(df_perfusion_v2.columns) & set(df_stroke_v2.columns)
print(common_col)
print(len(common_col))
```

On 67 and respectively 68 features, the two datasets share 51 features. Let's check the matches

```python
desc_common_el = df_desc_v1[df_desc_v1["SUBJECT NUMBER"].isin(common_col)][["SUBJECT NUMBER", "Variable description"]]
desc_common_el.set_index("SUBJECT NUMBER", inplace=True)
```

```python
unique_value_stroke = []
unique_value_perfusion = []
symmetric_difference = []

for col in common_col:
    unique_stroke = list(pd.unique(df_stroke_v2[col]))
    unique_perfusion = list(pd.unique(df_perfusion_v2[col]))
    unique_value_stroke.append(unique_stroke)
    unique_value_perfusion.append(unique_perfusion)
    symmetric_difference.append((set(unique_stroke).symmetric_difference(set(unique_perfusion))))
          
df_common_features_comparison = pd.DataFrame({'stroke':unique_value_stroke, 'perfusion':unique_value_perfusion, "symmetric_diff": symmetric_difference}, index=common_col)
df_common_features_comparison["desc"] = desc_common_el
df_common_features_comparison
```

Checking similarity
- atrial fibtrillation patient medical history: what is the /h ??
- insulin(yes_or_no): no 'yes' in the stroke dataset
- heart failure =chf /ifarction=-mi patient medical history
- numbness autonomic symptoms: '20' in stroke dataset
- group: DM/STROKE vs control
- stroke patient medical history: TIA in perfusion

## Uncommon features:

```python
uncommon_stroke = ['completed_visit_status',
                   'group2',
                   'htn_patient_medical_history',
                   'ldl calcmg/dl',
                   'race',
                   'rbc m/ul',
                   'rdw%',
                   'stroke group']
uncommon_perfusion = [ 'alb/cre mg/g (urine)',
                       'diabetes duration', 
                       'dmspecific family history',
                       'hb a1c%',
                       'medications',
                       'other 2',
                       'study',
                       'urine albumin mg/d',
                       'urine creat mg/dl']
```

Bio measures can be excluded: 
- stroke:    
    - 'rbc m/ul'
    - 'rdw%'
    - 'ldl calcmg/dl'
- perfusion: 
    - 'alb/cre mg/g (urine)'
    - 'hb a1c%'
    - 'urine albumin mg/d'
    - 'urine creat mg/dl'

```python
dump_features_stroke = ["rbc m/ul", 'rdw%', 'ldl calcmg/dl']
dump_features_perfusion = ['alb/cre mg/g (urine)', 'hb a1c%', 'urine albumin mg/d', 'urine creat mg/dl']
for el in dump_features_stroke:
    uncommon_stroke.remove(el)
    
for el in dump_features_perfusion:
    uncommon_perfusion.remove(el) 
```
